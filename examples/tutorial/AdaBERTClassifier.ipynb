# Teacher训练

```python
model = uf.BERTClassifier("../../ref/bert_config.json", "../../ref/vocab.txt", output_dir=".teacher")
print(model)
```

```python
X = [
    "天亮以前说再见，笑着泪流满面，去迎接应该你的，更好的明天", 
    "昙花若只一现，更要开的耀眼，别回头去拥有，属于你更好的世界", 
    "天亮以前说再见，让我留在今天，去保护我和你的，最好的夏天", 
    "情丝若水三千，只取一瓢眷恋，当你来过的纪念",
]
y = [1, 0, 1, 0]
```

```python
model.fit(X, y, total_steps=20)
```

```python
model.save()
```

# Student训练

```python
model = uf.AdaBERTClassifier("../../ref/bert_config.json", "../../ref/vocab.txt", init_checkpoint=".teacher", label_size=2)    # 配置文件依然使用teacher模型的
print(model)
```

```python
model.fit(X, y, total_steps=500, learning_rate=5e-2, optimizer="lamb")        # student模型较小，可以适当提高学习率
```

# 推理

```python
model.predict(X)
```

# 评分

```python
model.score(X, y)
```
