# AdaBERTClassifier

Teacher模型：`BERTClassifier`

```python
import uf

print(uf.__version__)
```

# Teacher训练

```python
model = uf.BERTClassifier("../../ref/bert_config.json", "../../ref/vocab.txt", output_dir=".teacher")
print(model)
```

```python
X = ["天亮以前说再见", "笑着泪流满面", "去迎接应该你的", "更好的明天"]
y = [1, 0, 1, 0]
```

```python
model.fit(X, y, total_steps=20)
```

```python
model.save()
```

# Student训练

```python
model = uf.AdaBERTClassifier("../../ref/bert_config.json", "../../ref/vocab.txt", init_checkpoint=".teacher", max_seq_length=16, label_size=2)    # 配置文件依然使用teacher模型的
print(model)
```

```python
model.fit(X, y, total_steps=500, learning_rate=5e-2)        # student模型较小，可以适当提高学习率
```

AdaBERT的模型设计使其在数据量较少时收敛困难

# 推理

```python
model.predict(X)
```

# 评分

```python
model.score(X, y)
```
